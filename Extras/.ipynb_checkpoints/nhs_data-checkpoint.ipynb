**Question 2:** What are the five locations with the highest number of appointments?

I used the groupby() function with 'sub_icb_location_name' column before aggregating the count of appointments, and the using the sort_values() to output the result based on the count of appointments in the descending order.  

#### Top 5 locations with the highest number and percentage of Appointments. 

```python
# Total appointments across all locations.
total_appointments = nc['count_of_appointments'].sum()
print(f"Total appointments across all locations: {total_appointments:,}")
print("\n")

# Determine the top five locations based on record count.
# Sort the highest appointment counts in descending order
# Sum of appointments per location.
highest_num_app = nc.groupby('sub_icb_location_name')[['count_of_appointments']].sum() \
.sort_values('count_of_appointments', ascending=False)

# Reset index so the location name becomes a regular column
highest_num_app = highest_num_app.reset_index()

# Calculate the percentage of each location.
highest_num_app['Percentage'] = (highest_num_app['count_of_appointments'] / 
                                  highest_num_app['count_of_appointments'].sum()) * 100

highest_num_app['Percentage'] = highest_num_app['Percentage'].apply(lambda x: f"{x:.2f}%")


# Print the top five Locations by appointment.
print("Top 5 Locations :", highest_num_app.head(5))
```

#### Total number and Percentage of Appointments by Service Settings


```python
# Count how many times each service setting appears (like number of rows)
count_by_setting = nc['service_setting'].value_counts().reset_index()
count_by_setting.columns = ['service_setting', 'count_of_service_setting']

# Sum total appointments by service setting
appointments_by_setting = nc.groupby('service_setting')['count_of_appointments'].sum().reset_index()
appointments_by_setting.columns = ['service_setting', 'total_appointments']

# Merge both into one DataFrame
summary_service_settings_app = pd.merge(count_by_setting, appointments_by_setting, on='service_setting')


# Calculate percentage of total appointments
total = summary_service_settings_app['total_appointments'].sum()
summary_service_settings_app['percentage'] = (summary_service_settings_app['total_appointments'] / total) * 100


# Sort by total appointments
summary_service_settings_app = summary_service_settings_app.sort_values(by='total_appointments', ascending=False)

# Format numbers with commas for readability
summary_service_settings_app['count_of_service_setting'] = summary_service_settings_app['count_of_service_setting'].apply(lambda x: f"{x:,}")
summary_service_settings_app['total_appointments'] = summary_service_settings_app['total_appointments'].apply(lambda x: f"{x:,}")
summary_service_settings_app['percentage'] = summary_service_settings_app['percentage'].apply(lambda x: f"{x:.2f}%")

# Print out the total appointments by each Service Setting
print(summary_service_settings_app)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert percentage back to float for plotting (remove % sign)
summary_service_settings_app['percentage_value'] = summary_service_settings_app['percentage'].str.replace('%', '').astype(float)

# Set plot style
sns.set_theme(style="whitegrid")

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
barplot = sns.barplot(
    data=summary_service_settings_app,
    x='percentage_value',
    y='service_setting',
    palette='Blues_d'
)

# Add chart labels
plt.title('Total Appointments by Service Setting (%)')
plt.xlabel('Percentage of Total Appointments')
plt.ylabel('Service Setting')

# Add % labels at the end of each bar
for i, row in summary_service_settings_app.iterrows():
    plt.text(row['percentage_value'] + 0.5, i, row['percentage'], va='center')

# Display the plot
plt.tight_layout()
plt.show()
```

#### A bar chart showing the total number of appointments by Service Settings

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert the percentage back to float for plotting
summary_context_type['percentage_value'] = summary_context_type['percentage'].str.replace('%', '').astype(float)

# Set Seaborn style
sns.set_theme(style="whitegrid")

# Create the horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(
    data=summary_context_type,
    x='percentage_value',
    y='context_type',
    palette='Greens_d'
)

# Add chart labels and title
plt.title('Total Appointments by Context Type (%)')
plt.xlabel('Percentage of Total Appointments')
plt.ylabel('Context Type')

# Add text labels to the bars
for i, row in summary_context_type.iterrows():
    plt.text(row['percentage_value'] + 0.5, i, row['percentage'], va='center')

# Tidy up layout and show the chart
plt.tight_layout()
plt.show()
```

#### Total number of Appointments with percentage by national Categories.

```python
# Count how many times each national category appears
count_by_category = nc['national_category'].value_counts().reset_index()
count_by_category.columns = ['national_category', 'count_of_national_category']

# Sum total appointments by national category
appointments_by_category = nc.groupby('national_category')['count_of_appointments'].sum().reset_index()
appointments_by_category.columns = ['national_category', 'total_appointments']

# Merge both into one DataFrame
summary_national_category = pd.merge(count_by_category, appointments_by_category, on='national_category')

# Calculate percentage of total appointments
total = summary_national_category['total_appointments'].sum()
summary_national_category['percentage'] = (summary_national_category['total_appointments'] / total) * 100

# Sort by total appointments (before formatting)
summary_national_category = summary_national_category.sort_values(by='total_appointments', ascending=False)

# Format numbers with commas and add '%' to percentage
summary_national_category['count_of_national_category'] = summary_national_category['count_of_national_category'].apply(lambda x: f"{x:,}")
summary_national_category['total_appointments'] = summary_national_category['total_appointments'].apply(lambda x: f"{x:,}")
summary_national_category['percentage'] = summary_national_category['percentage'].apply(lambda x: f"{x:.2f}%")

#  Display the result
print(summary_national_category)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Recalculate the percentage as float (if not already available)
summary_national_category['percentage_value'] = (
    summary_national_category['total_appointments']
    .str.replace(',', '')
    .astype(float)
)
total_appointments = summary_national_category['percentage_value'].sum()
summary_national_category['percentage_value'] = (summary_national_category['percentage_value'] / total_appointments) * 100
summary_national_category['percentage_value'] = summary_national_category['percentage_value'].round(2)

# Sort by percentage before plotting to match visual order
summary_national_category = summary_national_category.sort_values(by='percentage_value', ascending=False)

# Plot
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")
sns.set_context("talk")

ax = sns.barplot(
    data=summary_national_category,
    x='percentage_value',
    y='national_category',
    palette='Purples_d'
)

# Add accurate % labels using bar widths (patches)
for bar, perc in zip(ax.patches, summary_national_category['percentage_value']):
    width = bar.get_width()
    y = bar.get_y() + bar.get_height() / 2
    ax.text(width + 0.5, y, f"{perc:.2f}%", va='center')

# Labels and title
plt.xlabel('Percentage of Total Appointments')
plt.ylabel('National Category')
plt.title('Total Appointments by National Category (%)')

plt.tight_layout()
plt.show()
```

### Number of Appointments by Appointment Status ('Attended', Did not attend' & 'Unknown')

```python
import pandas as pd

# Count how many times each appointment status appears
count_by_status = ar['appointment_status'].value_counts().reset_index()
count_by_status.columns = ['appointment_status', 'count_of_appointment_status']

# Sum total appointments by appointment status
appointments_by_status = ar.groupby('appointment_status')['count_of_appointments'].sum().reset_index()
appointments_by_status.columns = ['appointment_status', 'total_appointments']

# Merge both into one summary DataFrame
summary_appointment_status = pd.merge(count_by_status, appointments_by_status, on='appointment_status')

# Calculate percentage of total appointments
total = summary_appointment_status['total_appointments'].sum()
summary_appointment_status['percentage'] = (
    summary_appointment_status['total_appointments'] / total
) * 100

# Sort by total appointments before formatting
summary_appointment_status = summary_appointment_status.sort_values(by='total_appointments', ascending=False)

# Format columns
summary_appointment_status['total_appointments'] = summary_appointment_status['total_appointments'].apply(lambda x: f"{x:,}")
summary_appointment_status['percentage'] = summary_appointment_status['percentage'].apply(lambda x: f"{x:.2f}%")

# Replace 'DNA' with 'Did Not Attend' for clarity
summary_appointment_status['appointment_status'] = summary_appointment_status['appointment_status'].replace({
    'DNA': 'Did Not Attend'
})

# Drop count of status column
summary_appointment_status = summary_appointment_status.drop(columns=['count_of_appointment_status'])

#  Rename columns for clean display
summary_appointment_status.columns = ['Appointment Status', 'Total Appointments', 'Percentage']

# Display the final summary
print(summary_appointment_status)
```

**Question 2:** Which service setting was the most popular (had the most appointments reported) for NHS North West London from 1 January to 1 June 2022?

```python
# First, filter the DataFrame for the required ICB location and date range
nc_subset = nc[
    (nc['sub_icb_location_name'] == 'NHS North West London ICB - W2U3Z') &
    (nc['appointment_date'] >= '2022-01-01') &
    (nc['appointment_date'] <= '2022-06-01')
]

# Group by service setting and calculate the total sum of appointments
service_total = nc_subset.groupby('service_setting')['count_of_appointments'].sum().reset_index()

# Sort the result by the number of appointments in descending order
service_total = service_total.sort_values('count_of_appointments', ascending=False)

# Identify the service setting with the highest number of appointments
max_service_setting = service_total.loc[service_total['count_of_appointments'].idxmax(), 'service_setting']

# Count the total number of appointments at the highest service setting.
app_total = service_total.loc[service_total['count_of_appointments'].idxmax(), 'count_of_appointments']

# Print the sorted service totals
print(service_total)

# Output the service setting with the highest number of appointments
# Output the service setting with the highest number of appointments and the count
print(f"Service Setting with the highest number of appointments: {max_service_setting}")
print(f"Number of appointments at this service setting: {app_total}")
```

**Question 3:** Which month had the highest number of appointments?

```python
# Ensure appointment_month is in datetime format
nc['appointment_month'] = pd.to_datetime(nc['appointment_month'])

# Group by month and sum appointments
month_app = nc.groupby(nc['appointment_month'].dt.to_period('M'))['count_of_appointments'].sum().reset_index()

# Convert Period to timestamp for display
month_app['appointment_month'] = month_app['appointment_month'].dt.to_timestamp()

# Sort by number of appointments in descending order
month_app = month_app.sort_values('count_of_appointments', ascending=False)

# Format numbers with commas
month_app['count_of_appointments'] = month_app['count_of_appointments'].apply(lambda x: f"{x:,}")

# Rename columns for clarity
month_app = month_app.rename(columns={
    'appointment_month': 'Month',
    'count_of_appointments': 'Total Appointments'
})

# Identify the row with the highest number of appointments
max_row = month_app.iloc[0]
highest_appointments = int(max_row['Total Appointments'].replace(',', ''))  # Convert back to int for formatting
most_popular_month = max_row['Month'].strftime('%B %Y')  # Format for readability


# Print the result
# Print insights
print(f"The highest number of appointments was {highest_appointments:,} in {most_popular_month}.\n")
print("\nMonths ranked by number of appointments (highest first):\n")
print(month_app)
```

### Total number of Daily Appointments against Daily Capacity Guideline of 1,200,000.

In order to understand if the NHS is exceeding or under utilising its resources such as Daily Appointment Capacity Guideline of 1,200,000, a dataframe was created with a dataframe was created with total number of daily appointments per month in the period between August 2021 and June 2022. Two new columns called Utilisation and % capacity (to calculate % of NHS daily capacity) were also added, and a visualisation in the form of a line chart was also created. 

```python
# Ensure appointment_month is in datetime format
ar['appointment_month'] = pd.to_datetime(ar['appointment_month'])

# Filter the data so that it starts from August 2021 onwards
filtered_data = ar[ar['appointment_month'] >= '2021-08-01']


# Group by month and calculate total appointments
monthly_df = filtered_data.groupby(filtered_data['appointment_month'].dt.to_period('M'))['count_of_appointments'].sum().reset_index()
monthly_df['appointment_month'] = monthly_df['appointment_month'].astype(str)
monthly_df['appointment_month'] = pd.to_datetime(monthly_df['appointment_month'])

# Add number of days per month
monthly_df['days_in_month'] = monthly_df['appointment_month'].dt.daysinmonth

# Calculate average daily appointments
monthly_df['avg_daily_appointments'] = monthly_df['count_of_appointments'] / monthly_df['days_in_month']

# Calculate % of NHS daily capacity used (1,200,000 appointments per day)
monthly_df['nhs_daily_capacity'] = 1200000
monthly_df['%_of_capacity'] = (monthly_df['avg_daily_appointments'] / monthly_df['nhs_daily_capacity']) * 100

# Round values
monthly_df['avg_daily_appointments'] = monthly_df['avg_daily_appointments'].round(0)
monthly_df['%_of_capacity'] = monthly_df['%_of_capacity'].round(1)

# Rename columns for display
summary_df = monthly_df.rename(columns={
    'appointment_month': 'Appointment Month',
    'count_of_appointments': 'Total Appointments',
    'avg_daily_appointments': 'Utilisation',
    '%_of_capacity': '% Capacity Used'
})

# View DataFrame separately
display(summary_df[['Appointment Month', 'Total Appointments', 'Utilisation', '% Capacity Used']])


# Plot the Chart 
plt.figure(figsize=(12, 6))
plt.plot(summary_df['Appointment Month'], summary_df['Utilisation'], marker='o', label='Avg Daily Appointments')
plt.axhline(y=1200000, color='r', linestyle='--', label='NHS Maximum Daily Capacity Guideline (1,200,000)')

# Format y-axis with commas instead of scientific notation
plt.ticklabel_format(style='plain', axis='y')

plt.title("NHS Average Daily Appointments vs Capacity", fontsize=14)
plt.xlabel("Appointment Month", fontsize=12)
plt.ylabel("Average Daily Appointments", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()

# Save the figure
fig.savefig("capacity_utilisation.png", dpi=300, bbox_inches='tight')

plt.show()
```

### Summary of the gathered Insights:

+ NHS has stayed comfortably within its available daily appointment capacity guideline of 1.2 million per day threshold. 

+ October 2021 to March 2022 showed a steady increase in daily appointments, peaking in November 2021 and March 2022, which could be due to flu season and winter illness spikes, suggesting appropriate service responsiveness.

+ Significant drops in December 2021 and April 2022 could likely be due to public holidays (e.g., Christmas, Easter) reduced clinic hours, staff leave or operational scaling-down

+ Daily appointments volume taper off after March, suggesting reduced seasonal demand. Therefore, appointments are in moderation during spring and summer seasons.

+ NHS is not exceeding its appointment delivery capacity, indicating that the infrastructure is sufficient for current demand levels.

### Looking at the trend of Missed Appointments.

To understand why appointments are missed and how the NHS can avoid it, four dataframes were created. The filtered_data (data between August 2021 and June 2022) was used. Furthermore, a variable for missed appointments (missed_appts) was created by filtering data only from 'Did Not Attend' appointment status. The first dataframe shows missed appointments over months, the second one shows missed appointments over month by hcp_types, the third one shows missed appointments over months based on appointment mode and the last one shows the same information based on Time between Book and Appointment. Finally, the charts were created using these dataframes.

```python
# Filter only missed appointments
missed_appts = filtered_data[filtered_data['appointment_status'] == 'DNA']

# 1. Missed Appointments Over Time 
missed_by_month = missed_appts.groupby(missed_appts['appointment_month'].dt.to_period('M'))['count_of_appointments'].sum().reset_index()
missed_by_month['appointment_month'] = pd.to_datetime(missed_by_month['appointment_month'].astype(str))

# 2. Missed Appointments by HCP Type 
missed_by_hcp = missed_appts.groupby([missed_appts['appointment_month'].dt.to_period('M'), 'hcp_type'])['count_of_appointments'].sum().reset_index()
missed_by_hcp['appointment_month'] = pd.to_datetime(missed_by_hcp['appointment_month'].astype(str))

# 3. Missed Appointments by Appointment Mode 
missed_by_mode = missed_appts.groupby([missed_appts['appointment_month'].dt.to_period('M'), 'appointment_mode'])['count_of_appointments'].sum().reset_index()
missed_by_mode['appointment_month'] = pd.to_datetime(missed_by_mode['appointment_month'].astype(str))

#  4. Missed Appointments by Time Between Booking and Appointment 
missed_by_time = missed_appts.groupby([missed_appts['appointment_month'].dt.to_period('M'), 'time_between_book_and_appointment'])['count_of_appointments'].sum().reset_index()
missed_by_time['appointment_month'] = pd.to_datetime(missed_by_time['appointment_month'].astype(str))
```

### Visualisations

```python
#  1. Total Missed Appointments Over Time 
plt.figure(figsize=(12, 6))
sns.lineplot(data=missed_by_month, x='appointment_month', y='count_of_appointments', marker='o')

# Format y-axis with commas instead of scientific notation
plt.ticklabel_format(style='plain', axis='y')
plt.title('Missed Appointments Over Time')
plt.xlabel('Appointment Month')
plt.ylabel('Missed Appointments')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)

# Save the figure
fig.savefig("missed_appts.png", dpi=300, bbox_inches='tight')
plt.show()
print("\n")

# 2. Missed Appointments by HCP Type 
plt.figure(figsize=(12, 6))
sns.lineplot(data=missed_by_hcp, x='appointment_month', y='count_of_appointments', hue='hcp_type', marker='o')

# Format y-axis with commas instead of scientific notation
plt.ticklabel_format(style='plain', axis='y')
plt.title('Missed Appointments by HCP Type Over Time')
plt.xlabel('Appointment Month')
plt.ylabel('Missed Appointments')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.legend(title='HCP Type')

# Save the figure
fig.savefig("missed_appts_hcp.png", dpi=300, bbox_inches='tight')
plt.show()
print("\n")

#  3. Missed Appointments by Appointment Mode 
plt.figure(figsize=(12, 6))
sns.lineplot(data=missed_by_mode, x='appointment_month', y='count_of_appointments', hue='appointment_mode', marker='o')

# Format y-axis with commas instead of scientific notation
plt.ticklabel_format(style='plain', axis='y')
plt.title('Missed Appointments by Appointment Mode Over Time')
plt.xlabel('Appointment Month')
plt.ylabel('Missed Appointments')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.legend(title='Appointment Mode')

# Save the figure
fig.savefig("missed_appts_mode.png", dpi=300, bbox_inches='tight')
plt.show()
print("\n")

#  4. Missed Appointments by Time Between Booking and Appointment 
plt.figure(figsize=(14, 6))
sns.lineplot(data=missed_by_time, x='appointment_month', y='count_of_appointments', hue='time_between_book_and_appointment', marker='o')
plt.title('Missed Appointments by Time Between Booking and Appointment')
plt.xlabel('Appointment Month')
plt.ylabel('Missed Appointments')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.legend(title='Time Gap')

# Save the figure
fig.savefig("missed_appts_tba.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Improved Hashtag Frequency Analysis
import re
import seaborn as sns
import matplotlib.pyplot as plt

# Initialize lists
tweets_text = []
tags = []

# Loop through each tweet
for tweet in tweets['tweet_full_text'].dropna():
    # Store tweet text (optional use)
    tweets_text.append(tweet)

    # Extract hashtags using regex (handles punctuation/formatting)
    hashtags = re.findall(r'#\w+', tweet.lower())
    tags.extend(hashtags)

# Create DataFrame for tweet text
tweets_text_df = pd.DataFrame({'tweet_text': tweets_text})

# Count hashtag frequency
tags_series = pd.Series(tags).value_counts().reset_index()
tags_series.columns = ['hashtag', 'count']

# Display top 30
print("Top 30 Hashtags:")
print(tags_series.head(30))

# Plot top 20 hashtags
top_tags = tags_series.head(20)
plt.figure(figsize=(12, 8))
sns.barplot(data=top_tags, x='count', y='hashtag', palette='viridis')
plt.title('Top 20 Most Frequent Hashtags')
plt.xlabel('Frequency')
plt.ylabel('Hashtag')
plt.tight_layout()
plt.show()
```
